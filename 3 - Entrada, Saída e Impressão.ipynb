# **Entradas, saídas e impressões**

---


### O que são entradas e saídas?
**Entradas**

Os dispositivos de entrada são os que inserem informações em um sistema a partir de uma fonte externa.

*Exemplo: WebCam, Scanner, leitor de cod de barras, etc...*


---

**Saídas**

Os dispositivos de saídas devolvem para o mundo externo as informações processadas pelos dispositivos.

Exemplo: Fone de ouvido, impressora, projetor, etc...

# **Como funcionam as entradas em pseudo-linguagem?**

Resposta: atravez de comandos, que podem variar de linguagem pra linguagem. Mas... no geral, a lógica é a seguinte:

*Primeiro a palavra LEIA(em maiúsculo mesmo) seguido do nome da variável entre parênteses( ), segue exemplo prático:


```
Inicio-metodo


   INTEIRO idade
   STRING sexo
   REAL salario

   LEIA(idade)

   LEIA(sexo)

   LEIA(salario)


fim-metodo

```
---

E é dessa forma que você capta dados dos usuários e os armazena em algum lugar, através das entradas.

# Como transformar esses comandos para Python?
Em Python o comando de entrada que vamos utilizar é o **'input'** e como nos usamos o input? Muito simples, você cria a variável seguida de " = " + "input", da seguinte forma:


````
idade = int(input("Digite a sua idade: "))
sexo = str(input("Digite o sexo: "))
salario = float(input("DIgite o seu salário: "))
````

# **Saídas**

Em algoritimos/pseudo-linguagem, as saídas são representadas basicamente pela palavra ESCREVA (em maiúsculo mesmo). O comando ESCREVA não faz nada mais do que escrever alguma coisa na tela, óbvio. 



```
inicio-metodo


    ESCREVA (“Oi mundo”)
    ESCREVA (idade);
    ESCREVA ("idade");
    ESCREVA (“O Sexo é: ”, sexo)
    

Fim-metodo
```




## **Como funciona o comando ESCREVA em Python?**

O ESCREVA, como ja vimos, é um tipo de saída, ou seja, um meio que o computador usa para se comunicar com o usuário.

Em Python esse comando se transforma em "**print**", isso mesmo, um simples e puro **print**.

veja como aplicar o print em um algoritimo em Python:



````
idade = int(input("Digite a sua idade: "))
sexo = str(input("Digite o sexo: "))
salario = float(input("DIgite o seu salário: "))

print("Sua idade é",idade,"e seu salário é", salario)
````

# **Observação importante**
Poise, a vida não é tão simples e na maioria das vezes vocês terão que declarar qual o *TIPO* de entrada que se trata aquela variável, se é uma **String**('Maricleudes'), **Inteiro**(5, -10, 1000), **Número Real** (1,25454), **Booleano**(true ou false), etc...


Em pseudo-linguagem, usamos os seguintes tipos de dados:

* INTEIRO = (int é só em python, pseudo-linguagem é INTEIRO)
* REAL = (serve para números com casas decimais, tanto Float como Double))
* STRING = (cadeia de caracteres)
* BOOLEANO = (sentenças com dois operadores lógicos, V ou F)



**Como funcionam esses tipos em Python?**


1º Declarar o nome da variável seguido de igualdade, exemplo : num1 = 

2º Declarar o tipo da variável, string(**str**), inteiro(**int**), float(**float**), Boolean(**bool**).

3º invocar o input >>> int(input()).



### **Segue exemplo prático:**



````
tem_filhos = bool(input("Você tem filhos? "))
print(tem_filhos)
````
