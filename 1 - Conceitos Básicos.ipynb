Por exemplo, para se trocar o pneu de um carro precisamos:


```
Ferramentas:

- Step
- Carro
Como fazer:

- Levante o carro perto da roda a ser trocada
- Afrouxar a roda do pneu do carro
- Troque o pneu antigo pelo novo
- Coloque os parafusos
```

Para se trocar uma lâmpada, precisamos:


```
Ferramentas:
- Nova lâmpada
- Escada

Como fazer:
- Desligar o disjuntor
- Posicionar a escada em baixo da lâmpada
- Segurara a lâmpada nova na mão
- Subir a escada
- Desenroscar a antiga lâmpada no sentido horário
- Enroscar a nova lâmpada

```

Com isso, esbarramos na seguinte definição de algoritmo:

### "Um algoritmo é uma sequência de passos **bem definida** e **finita**, com o objetivo de se solucionar um determinado problema"

Atentem para o *bem definida* e *finita*:


*   Bem definida: os passos realizados no algoritmo precisam ser muito claros quanto ao seu objetivo;
*   Finita: em algum momento da vida, este algoritmo precisa terminar e retornar algo para o usuário.

Utilizando uma linguagem qualquer (como Java, Python, ou mesmo o próprio português), podemos representar a sequência de passos executados, como se estivéssemos dando ordens ao computador. Os princípios básicos de um algoritmo consiste em:

* Todo algoritmo tem um início
* Todo algoritmo tem um fim
* Todo algoritmo precisa retornar algo para o usuário. Um algoritmo que não retorna nada não tem utilidade nenhuma

Os elementos que compõem
* Variáveis: Armazenam informações/valores na memória do computador
* Sintaxe: Regras de como escrever um código legível
* Compilador: Responsavel por "traduzir" o algoritmo para linguagem nativa da máquina


Passos para uma soma:
```
- Valor A
- Valor B
- Expreção: A + B
- Resultado = A + B
```


````
a = 5
b = 10
resultado = a + b
print(resultado)
````
