# **Estruturas Condicionais**

Qualquer programa de computador pode ser escrito
combinando-se os três tipos básicos de estruturas de
programação:

*   Estrutura Sequencial
*   Estrutura Condicional
*   Estrutura de Repetição

## **Estrutura Sequencial**
É a estrutura de programação mais simples. 
Os fluxos de comando do algoritmo segue a mesma sequência linear da nossa escrita comum:


*   De cima para baixo
*   Da esquerda para a direita
Sua estrutura básica é a repetição de **comando sequencial**:

```
Ação1;
```

```
Ação1;
Ação2;
Ação3;
Ação4;
Ação5;
```
Exemplo:

````
# Tirar a média das notas
n1, n2, n3 = 8, 9, 10
media = (n1+n2+n3)/3
print(media)
````

## **Estrutura Condicional**
A estrutura **SE...SENÃO** é a mais simples estrutura de controle em linguagem de programação. Esta estrutura permite executar um entre **vários blocos de instruções**.

O controle de qual bloco será executado é dado por uma condição, **expressão lógica ou numérica**. Esta estrutura pode se apresentar de dois modos diferentes: condicional simples ou composta.

No final a condição terá um valor **Verdadeiro** ou **Falso**

![alt text](http://www.vichinsky.com.br/Cbasico/Image1.gif)

Exemplo de estrutura condicional simples:

````
# Verificar se o aluno aprovou
nota = int(input("Nota: "))
if (nota>=5):
  print("Parabéns, vc é crânio!!")
````

Condicional composta: Decisão de dois blocos. Também é possível escrever uma estrutura que execute um entre dois blocos de instruções.

![alt text](http://www.vichinsky.com.br/Cbasico/Image2.gif)

Exemplo:

````
# Verificar se o aluno aprovou, se não aprovou mande uma mensagem
nota = int(input("Nota: "))
if (nota>=5):
  print("Parabéns, vc é crânio!!")
else:
  print("Tururuuuu")
````

Para que os **blocos de código** funcionem de acordo com a condição, temos que deixar o nosso código **identado** pois se não isso pode gerar erros.

Para identar aperte o *tab*

![alt text](https://4.bp.blogspot.com/-ZViITKaKIlE/XM23goGh7pI/AAAAAAAAHwo/DsvOYCX6EeggBhE603zuyOl57x_dWUsIgCLcBGAs/s400/if.png)

````
nota = float(input("Nota: "))
if (nota>=8):
  print("Parabéns, vc é crânio!!")
elif (nota>=5):
  print("Você passou meu amigo.")
elif (nota<5):
  print("Tururuuuu")
````

Vamos agora ver um exemplo utilizando os **operadores lógicos e aritméticos**, transformando a frase a seguir em um algoritmo. 

Primeiro utilizando o operador lógico AND:

**"Vou ao cinema SE tiver tempo E dinheiro".**



````
print("Miga sua loka:")
tempo = str(input("Você tem tempo?"))
dinheiro = float(input("Quanto dinheiro você tem?"))
if(tempo=='sim' and dinheiro>=20):
  print("Bora ir para o cine")
else:
  print("Que peninha")
````

Tabela verdade AND (MULTIPLICAÇÃO 0 E 1)
> V AND V = V //1 X 1 = 1

> V AND F = F //1 X 0 = 0

> F AND V = F //0 X 1 = 0

> F AND F = F //0 X 0 = 0

E com o operador lógico OR? 

Segue o fio:

**"Irei pra balada SE meus amigos estiverem lá OU se meu crush estiver lá".**

````
print("Bora na balada?")
amigos = str(input("Os amigos estão na balada? "))
crush = str(input("O @ está na balada? "))
if(amigos=='sim' or crush=='sim'):
  print("Já estou lá")
else:
  print("Não vou não")


````

Tabela verdade OR (SOMA 0 E 1)
> V OR V = V //1 + 1 = 1

> V OR F = V //1 + 0 = 1

> F OR V = V //0 + 1 = 1

> F OR F = F //0 + 0 = 0

Pronto!
