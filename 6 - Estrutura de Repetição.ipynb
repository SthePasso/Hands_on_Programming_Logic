# Estrutura de Repetição

---


**O que é estrutura de repetição?**

É a estrutura em **algoritmos que permite executar mais de uma vez um conjunto de comandos**, ou bloco de comandos, de acordo com uma condição ou com um contador. 

Temos 3 tipos de estruturas de repetição:

- Para...faça (for)
- Enquanto...faça (while)

**O que é um contador e o que é um acumulador**
- **Contadores**

Um contador é utilizado para contar o número de vezes que um evento (uma instrução) ocorre, ou seja, contar a quantidade de vezes que uma instrução é executada. 
```
Forma Geral: variavel = variavel + constante 

Exemplo: X = X + 1 
```
Explicação: um contador é uma variável (qualquer) que recebe ela mesma mais um valor (uma constante), no caso do exemplo acima, a variável X está recebendo o valor dela mesma mais 1. 

Normalmente a constante que será somada no contador é o valor 1, para contar de 1 em 1, mas pode ser qualquer valor, como por exemplo 2, se quisermos contar de 2 em 2.
```
Algoritmo "Teste de Variável"
Declaração das variáveis
   nome : Texto
   contador : Inteiro(0)
Inicio
   nome <- “João”
   contador = 1+contador
   imprimir(nome)

   nome <- “Maria”
   contador = 1+contador
   imprimir(nome)
Fim
--------------------------------------------------------------------
Memória   |   Tipo Var   |   Nome   |   Conteúdo   |   Interação
  123     |    Texto     |   nome   |    João      |      1
               Texto     |   nome   |    Maria     |      2
  124     |   Inteiro    | contador |      0       |      1
              Inteiro    | contador |      1       |      2
              Inteiro    | contador |      2       |      0
  125     |              |          |              |      
  126     |              |          |              |      
```
 

Observações dos Contadores: 
> 1. A **variável (do contador) deve possuir um valor inicial conhecido**, isto é, ela deve ser inicializada. Normalmente inicializa-se a variável do contador com zero, ou seja, zera-se a variável antes de utilizá-la. Para zerar uma variável basta atribuir a ela o valor zero: VARIÁVEL 0 

> 2. **A constante (que é geralmente o valor 1)** determina o valor do incremento da variável (do contador), ou seja, o que será somado (acrescido) a ela. 


- **Acumuladores (ou Somadores)**

Um acumulador também é conhecido como somador.

Forma Geral: variavel1 = variavel1 + variavel2 

Exemplo: x = x + y 

Explicação: um acumulador (somador) é uma variável (qualquer) que recebe 
ela mesma + uma outra variável. No caso do exemplo acima, a variável X está recebendo o valor dela mesma mais o valor da variável Y. A variável Y representa o valor a ser somado, acumulado na variável X. 

Observações dos Acumuladores: 

>1. A **variável1 (do acumulador) deve possuir um valor inicial conhecido**, isto é, ela deve ser inicializada. Normalmente inicializa-se a variável do acumulador com zero, ou seja, zera-se a variável antes de utilizá-la. Para zerar uma variável basta atribuir a ela o valor zero: variavel1 = 0 

>2. A **variável2 indica o valor a ser acumulado**, somado e armazenado na variável1

## Enquanto...faça
**While** – Esta instrução é usada quando **não sabemos quantas vezes** um determinado bloco de instruções precisa ser repetido. Com ele, a execução das instruções vai continuar até que uma condição seja verdadeira. A condição a ser analisada para a execução do laço de repetição deverá retornar um valor booleano.


![alt text](https://i1.wp.com/upload.wikimedia.org/wikipedia/commons/thumb/4/43/While-loop-diagram.svg/440px-While-loop-diagram.svg.png?resize=310%2C392&ssl=1)

- Sintaxe:
```
.enquanto <condicão> faca
.  //bloco de codigo
.  fimenquanto
```
- Legenda
```
<condicao> : operador relacional ou lógico
```
- Pseudocódigo:
```
.i : inteiro(0)
.enquanto i<10 faca
.  imprima(i)
.  i=i+1
.fimenquanto
```
Perceba que, **somente se a condição for verdadeira o corpo do laço de repetição serão executados**. Portanto, o conteúdo será repetido até que esta condição não seja mais verdadeira.

Exemplo: Receba um número do usuário e o incremente(contador) e imprima enquanto for menor que 10.


````
i = 0
while (i<8):
  i = i+2  
  print(i)# 2 4 6 8
````

![alt text](https://pbs.twimg.com/media/EAXtdCwXkAAwi-n.jpg)

Exemplo: Receba um número do usuário, o eleve ao quadrado e imprima enquando for menor que 100.

````
numero = int(input("Digite um número: "))
while numero<100:
  print(numero)
  numero = numero**2
````

Exemplo 2: Escreva um programa que receba um número do usuário enquanto este número for diferente do número da sorte.

````
sorte = 77
numero = 0
while (sorte != numero):
  numero = int(input("Digite um número: "))
print("Parabéns")
````

## Para...faça
Este laço de repetição está associado a um contador. Seu uso é mais indicado para situações onde o número de iterações de um laço é bem definido. Por exemplo, escreva os números de 0 a 10.

![alt text](https://lh6.googleusercontent.com/proxy/BFlfiJMqY3Ednno03jITtphUuRzTfqxZFbXyaOzBPYgVyyQeo-OKBAu21Y6uM3t10Y-CdUDrwrYjO1hOd01xE8v7rOv576Zra1YRkfS6v6TWwuIAPli43yWXHG-MYAOgRtq-2w=s0-d)
- Sintaxe:
```
.para <var> de <inicio> ate <fim> faca
.  //bloco de codigo
.fimpara
```
- Legenda
```
var : variável inteira
inicio : numero inteiro de inicio
fim: numero inteiro fim
```
- Pseudocódigo:
```
.i : inteiro
.para i de 0 ate 10 faca
.  imprima(i)
.fimpara
```

Exemplo: Imprima todos os número pares entre 0 e 10.

````
for i in range(0,10):
  if(i%2==0):
    print(i)
````

**Conclusão** As repetições permitem a criação de códigos com a lógica mais apurada e com tamanhos mais redusidos e mais confiáveis.

## **Exercícios**

> 1. Receba um salário e o último aumento de salário que teve no ano de 2019 e imprima esse e todos os aumentos enquanto forem menores que 500.

````
salario = float(input("Digite seu salário: "))
aumento = float(input("Digite seu aumento em 2019: "))
acumulador = aumento
while acumulador < 500:
  print("Salario atual: ",salario+acumulador)
  print("Aumento: ",acumulador)
  acumulador = acumulador+aumento
````

> 2. Escreva um algoritmo para ler a nota de 10 alunos e contar quantos foram aprovados, sendo que, para ser aprovado, a nota deve ser maior ou igual a 6,0. Escrever o número de aprovados. (ATENÇÃO: Se quiséssemos contar o número de reprovados também, no exemplo acima, o que deveria ser feito?)

````
cont = 0
for i in range(0,10):
  nota = float(input("Digite a nota: "))
  if(nota>=6):
    cont+=1 #cont=cont+1
print(cont, "alunos foram aprovados")
````

## Aula de Revisão

Hoje vamos aprender uma técnica muito boa para verificar algoritmos de repetição, o **Teste de Mesa**.

Teste de Mesa é uma simulação da execução de um algoritmo, utilizando o processador mais potente já conhecido na história: a sua cabeçola.

Ele é realizado de forma manual, geralmente feita no papel, estilo raiz, pra criar caráter.

![alt text](https://media1.giphy.com/media/l41m04gr7tRet7Uas/giphy.gif)

Mas como fazer um teste de mesa?

Vamos aprender na prática, codificando e testando um algoritmo que calcula os números Fibonacci. Fibo o quê?

Na matemática, a **Sucessão de Fibonacci** (também Sequência de Fibonacci), é uma sequência de números inteiros, começando normalmente por 1, na qual, cada termo subsequente corresponde à soma dos dois anteriores. 


![alt text](https://media.giphy.com/media/gDUpfPXqShIXK/giphy.gif)

**Teste de Mesa**

> 1o passo: Escreva as variaveis

> 2o passo: Coloque o conteudo delas de acordo com o código

Vamos agora implementar um algoritmo em python que recebe como parâmetro um número inteiro. O algoritmo deve imprimir a quantidade de termos da série Fibonacci, de 0 até este número. Por exemplo, se a entrada for 4, deve imprimir: 0 1 1 2 3

````
anterior = 0
proximo = 0
ate = int(input())
while(proximo < ate):
  print(proximo)
  proximo = proximo + anterior
  anterior = proximo - anterior
  if(proximo == 0):
    proximo = proximo + 1
````

**Faça um programa que dado dois números inteiros N1 e N2 de tal forma que N1 seja multiplicado de 0 até N2.**



````
n1 = int(input())
n2 = int(input())

for i in range(0, n2+1):
  print("{}x{}: {}".format(i, n1, i*n1))
  
````

**Receba um número do usuário e imprima seu fatorial**

````
numero = int(input())
cont = 0
fatorial = 0
while(cont <= numero):
  if(cont==0):
    fatorial += 1
  else:
    fatorial *= cont
  cont+=1

print(fatorial)
````
